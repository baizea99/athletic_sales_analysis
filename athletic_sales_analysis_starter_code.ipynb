```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_path = Path('Resources/athletic_sales_2020.csv')
sales_2021_path = Path('Resources/athletic_sales_2021.csv')

sales_2020_df = pd.read_csv(sales_2020_path)
sales_2021_df = pd.read_csv(sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

# Concatenate the DataFrames along the rows
combined_sales_data = pd.concat([sales_2020_df, sales_2021_df])

# Reset the index
combined_sales_data = combined_sales_data.reset_index(drop=True)
combined_sales_data
```

```python
# Check if any values are null.
combined_sales_data.isnull().any().any()
```

```python
# Check the data type of each column
combined_sales_data.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_data['invoice_date'] = pd.to_datetime(combined_sales_data['invoice_date'], format='%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_data.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

# Grouping by region, state, and city and counting the number of products sold
top_products_sold = combined_sales_data.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index(name='Total_Products_Sold')


# Show the top 5 results.
top_5_products_sold = top_products_sold.sort_values(by='Total_Products_Sold', ascending=False).head(5)
top_5_products_sold                  
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table = pd.pivot_table(combined_sales_data, 
                             values='units_sold', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'})


# Show the top 5 results.
top_products_sold = pivot_table.sort_values(by='Total_Products_Sold', ascending=False).head(5)

top_products_sold
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

# Show the total sales for the products sold for each region, state, and city.
combined_sales_data['total_sales'] = combined_sales_data['units_sold'] * combined_sales_data['price_per_unit']
total_sales_by_location = combined_sales_data.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

total_sales_by_location = total_sales_by_location.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_total_sales = total_sales_by_location.sort_values(by='Total Sales', ascending=False).head(5)

top_total_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table = pd.pivot_table(combined_sales_data, 
                             values='total_sales', 
                             index=['region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
top_total_sales = pivot_table.sort_values(by='Total Sales', ascending=False).head(5)
top_total_sales
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_by_location = combined_sales_data.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
total_sales_by_location = total_sales_by_location.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
top_total_sales = total_sales_by_location.sort_values(by='Total Sales', ascending=False).head(5)
top_total_sales
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_table = pd.pivot_table(combined_sales_data, 
                             values='total_sales', 
                             index=['retailer', 'region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_total_sales = pivot_table.sort_values(by='Total Sales', ascending=False).head(5)
top_total_sales
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_sales = combined_sales_data[combined_sales_data['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales_by_location = womens_athletic_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

womens_footwear_sales_by_location = womens_footwear_sales_by_location.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
top_womens_footwear_sales = womens_footwear_sales_by_location.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_womens_footwear_sales
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table = pd.pivot_table(womens_athletic_footwear_sales, 
                             values='units_sold', 
                             index=['retailer', 'region', 'state', 'city'], 
                             aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table = pivot_table.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
top_womens_footwear_sales = pivot_table.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_womens_footwear_sales
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table = pd.pivot_table(womens_athletic_footwear_sales, 
                             values='total_sales', 
                             index='invoice_date', 
                             aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})


# Show the table.
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

daily_sales = pivot_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales_sorted = daily_sales.sort_values(by='Total Sales', ascending=True)

daily_sales_sorted
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = pivot_table.resample('W').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales_sorted = weekly_sales.sort_values(by='Total Sales', ascending=True)
weekly_sales_sorted
```
